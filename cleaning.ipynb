```python
import pandas as pd
```

```python
df = pd.read_csv(r"Data/Superstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.columns
```

```python
# standardizing columns
df.columns = df.columns.str.strip().str.lower().str.replace (" ","_").str.replace("-","_")
```

```python
df.columns.tolist()
```

```python
#fixing datatypes
df['order_date']= pd.to_datetime(df['order_date'])
df['ship_date']=pd.to_datetime(df['ship_date'])
```

```python
df.dtypes
```

```python
df['postal_code'] = df['postal_code'].astype(str)
```

```python
df.dtypes
```

```python
#checking for null values
df.isnull().sum()
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#how long did delivery take
df['delivery_date'] = (df['ship_date']-df['order_date']).dt.days
```

```python
#Revenue flags
median_sales = df['sales'].median()
df['order_value_tier'] = df['sales'].apply(
    lambda x: "High value" if x >= median_sales else "Low value"
)
```

```python
#Extracting year and month for trend analysis
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.strftime("%b")
```

```python
df.head()
```

```python
df.to_csv(r"Data/cleaned_superstore.csv")
```
